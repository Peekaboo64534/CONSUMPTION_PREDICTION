```python

import os
import sys
import pandas as pd
import re
```


Locate all files in the directory

```python
directory = "/Users/peekaboo645/code/Peekaboo64534/project_name/CONSUMPTION_PREDICTION/"
file_names = []

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_names.append(filename)
```

```python

file_names
```

```python
def transform_file(file_name):
    
    df = pd.read_csv(file_name, header=None)
    
    #rename columns
    df.columns = ['datetime', 'gas']
    
    #convert to dt object
    df['datetime'] = pd.to_datetime(df['datetime'])

    # Aggregate the data daily
    daily = df.resample('D', on='datetime').sum()

    #create new column kWh
    daily['gas (kWh)'] = daily['gas']/1000
    daily.reset_index(inplace=True)

    daily.drop(['gas'], axis=1, inplace=True)

     # Extract homeid from the filename using regular expression
    homeid = re.search(r'\d+', file_name).group()
    
    # Add homeid
    daily['homeid'] = homeid

    return daily
```


## export dataframe to csv file and delete original file

```python
# export dataframe to csv file 
def export_file(df, file_name):
    df.to_csv('data_transformed/' + file_name, index=False)
    print('File exported successfully!')

   
```

```python
# loop through all files in the directory and transform them
for file_name in file_names:

    df = transform_file(file_name)
    export_file(df, file_name)
```
