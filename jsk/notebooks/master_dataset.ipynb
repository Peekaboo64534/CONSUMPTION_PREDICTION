```python
import pandas as pd
```

### Import and combine target datasets

```python
df_gas = pd.read_csv('../bigframe_gas.csv', index_col=0)
df_electricity = pd.read_csv('../data_jsk/master_data_electricity.csv')
```

```python
df_gas
```

```python
df_electricity
```

```python
# merge both dataframes on the date column by homeid
df_merged = pd.merge(df_gas, df_electricity, on=['datetime', 'homeid'])
df_merged = df_merged[['datetime', 'homeid','gas (kWh)', 'Consumption (kWh)']]

# new  column with total energy consumption
df_merged['Total (kWh)'] = df_merged['gas (kWh)'] + df_merged['Consumption (kWh)']
df_merged
```

### Import and combine features dataset with target dataset

```python
df_features = pd.read_csv('../data_jsk/macro_features.csv')

# show all columns
pd.set_option('display.max_columns', None)

df_features
```

```python
df_features.columns
```

```python
# select the needed features
selected_features = ['time',
                     'temp_mean',
                     'temp_min',
                     'temp_max',
                     'location_id',
                     'location_name',
                     'homeid',
                    #  'location',
                     'residents',
                     'hometype',
                    #  'equivalised_income',
                     'occupied_days',
                     'occupied_nights',
                     'entry_floor',
                     'income_band_num',
                    #  'start_date',
                    #  'end_date',
                    #  'days_between',
                     'electric_appliance',
                     'gas_appliance',
                     'total_area',
                     'room_count',
                     'windowsopen']

df_features_selected = df_features[selected_features].copy()

#rename columns time to datetime
df_features_selected.rename(columns={'time': 'datetime'}, inplace=True)
```

```python
df_features_selected
```

```python
# Merge df_merged with df_features_selected on the datetime and homeid columns
df_final = pd.merge(df_merged, df_features_selected, on=['datetime', 'homeid'])
```

```python
df_final
```

```python
# save dataset as csv
df_final.to_csv('../data_jsk/final_dataset_macro.csv', index=False)
```

```python
df_final['homeid']
```

```python
df_humidity = pd.read_csv('../data_jsk/humidity_data.csv')
df_humidity
```

```python
# merge humidity data with final dataset
df_final = pd.merge(df_final, df_humidity, on=['datetime', 'homeid'])
df_final
```

```python
df_temp_internal = pd.read_csv('../data_jsk/temperature_data.csv')
df_temp_internal
```

```python
# merge temperature data with final dataset
df_final = pd.merge(df_final, df_temp_internal, on=['datetime', 'homeid'])
df_final
```

```python
# save dataset as csv
df_final.to_csv('../data_jsk/final_dataset.csv', index=False)
```
