```python
import pandas as pd
```

```python
df_main = pd.read_csv('../../final_dataset.csv')
df_person = pd.read_csv('../df_final_person.csv', index_col=0)
```

```python
# merge the two dataframes
df_final = pd.merge(df_main, df_person, on='homeid', how='left')
df_final
```

```python
df_final.columns
```

```python
# rename consumption column
df_final.rename(columns={
                    'Consumption (kWh)': 'Electricity (kWh)',
                    'gas (kWh)': 'Gas (kWh)',
                         }, inplace=True)
```

```python
df_final['mean_h'] = df_final['mean_h']/10
df_final['min_h'] = df_final['min_h']/10
df_final['max_h'] = df_final['max_h']/10
df_final['mean_t'] = df_final['mean_t']/10
df_final['min_t'] = df_final['min_t']/10
df_final['max_t'] = df_final['max_t']/10
```
