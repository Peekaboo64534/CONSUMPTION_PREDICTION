```python
import pandas as pd

# Load the data
filename = "/content/home240_hall2226_sensor14265_gas-pulse_gas.csv"
data2 = pd.read_csv(filename, header=None)
data2.columns = ['datetime', 'gas']
data2['datetime'] = pd.to_datetime(data2['datetime'])
```

```python
# Aggregate the data daily
daily = data2.resample('D', on='datetime').sum()


```

```python
daily['gas (kWh)'] = daily['gas']/1000
```

```python
daily.describe()
```

```python
daily
```

```python

# Calculate the average daily gas consumption in kWh
average_daily_kwh = daily['gas (kWh)'].mean()

# Display the results
print(f"Average daily gas consumption: {average_daily_kwh:.2f} kWh")



daily[['gas (kWh)']]
```

```python
import plotly.express as px




fig = px.bar(daily.reset_index(), x='datetime', y='gas (kWh)',
             title='Daily Gas Consumption in kWh',
             labels={'datetime': 'Date', 'gas (kWh)': 'Gas Consumption (kWh)'},
             template='plotly_white')

# Show the interactive plot
fig.show()
```
